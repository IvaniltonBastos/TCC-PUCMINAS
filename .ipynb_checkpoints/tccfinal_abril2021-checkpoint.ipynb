Qual o comportamento da série temporal do total de vendas mês/ano no periodo de 2008 a 2020?

```python
    # criação de um modelo de machine learn para previsão de vendas anuais 
```

```python
#verificando o tipo dos dados para realizar o treinamento dos modelos 

dfEmplacamentosAnuais.dtypes
```

```python
display(dfEmplacamentosAnuais)
```

```python
dfEmplacamentosAnuais.drop(['1ºTrimestre','2ºTrimestre','3ºTrimestre','4ºTrimestre'], axis=1, inplace=True)
```

```python
display(dfEmplacamentosAnuais)
```

```python
x = dfEmplacamentosAnuais.drop('Total de Vendas', axis='columns')
y = dfEmplacamentosAnuais['Total de Vendas']
```

```python
#treinando nosso modelo com 30% da base de dados e separando os 20% restantes para testar o modelo

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
rf_reg =  LinearRegression()
rf_reg.fit(x_train, y_train)
```

```python
# Testes do Machine Learn
test_pred_lin = lin_reg.predict(x_test)
test_pred_rf = rf_reg.predict(x_test)
```

```python
r2_lin = metrics.r2_score(y_test,test_pred_lin)
rmse_lin = np.sqrt(metrics.mean_squared_error(y_test, test_pred_lin))
print(f'R² da Regressão Linear: {r2_lin}')
print(f'RSME da Regressão Linear: {rmse_lin}')
r2_rf = metrics.r2_score(y_test, test_pred_rf)
rmse_rf =np.sqrt(metrics.mean_squared_error(y_test, test_pred_rf))
print(f'R² do Random Forest:{r2_rf}')
print(f'RSME do Random Forest:{rmse_rf}')

```

```python
#Análise dos dados da Previsão
df_resultado = pd.DataFrame()
df_resultado['y_teste'] = y_test
df_resultado['y_previsao_rf'] = test_pred_rf
df_resultado['y_previsao_lin'] = test_pred_lin
df_resultado['Ano'] = dfEmplacamentosAnuais['Ano']
df_resultado['Total de Vendas'] = dfEmplacamentosAnuais['Total de Vendas'] 
df_resultado = df_resultado.reset_index(drop=True)
fig = plt.figure(figsize=(15,5))
#sns.lineplot(data=df_resultado,)
df_resultado.plot(kind='line', figsize=(18,5),title='Vendas Anuais', grid='True')

plt.show()


 
```

# Conclusão

Através da técnica de Scrapping foi obtida a série temporal para o emplacamento mensal dos últimos anos.O emplacamento mensal reflete as vendas mensais de veículos novos no Brasil.

Após as análises realizadas conclui que o melhor ano para a venda de veículos novos no Brasil foi o de 2012, enquanto em que 2016 apresentou o menor índice de vendas.
Observando o gráfico Trimestral, pude perceber que o melhor Trimestre para a venda de veículos é os 4º Trimestres e os 1º Trimestres são os trimestres com menor índice de vendas. Foi observado também, que em 2015, houve uma inversão de vendas ao longo dos trimestres.
